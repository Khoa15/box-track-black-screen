```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("apple.jpg", cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
blank_img = np.zeros((1000, 998, 3))
blank_img.shape
```

```python
x, y = 100, 200
w, h = 800, 710
startPoint = (x, y)
endPoint = (x + w, y + h)
cv2.rectangle(img, startPoint, endPoint, (0, 0, 0), 3)
plt.imshow(img)
plt.show()
```

```python
blank_img[y: y+h,x: x+w] = img[y: y+h, x: x+w]
plt.imshow(img[y: y+h, x: x+w])
plt.show()
```

```python
plt.imshow(blank_img)
plt.show()
```
